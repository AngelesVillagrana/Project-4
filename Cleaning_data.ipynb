```python
import seaborn as sns
import matplotlib.pyplot as plt

# AnÃ¡lisis de la variable de abandono en funciÃ³n de las variables demogrÃ¡ficas
demographic_columns = ['Gender_M', 'Marital_Status_Married', 'Education_Level_Graduate', 'Income_Category_$60K - $80K']

for col in demographic_columns:
    sns.countplot(x='Attrition_Flag', hue=col, data=BankChurners_data_scaled_df)
    plt.title(f'Attrition Flag by {col}')
    plt.show()
```
